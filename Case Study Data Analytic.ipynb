```python
loaddataset['last_review'] = pd.to_datetime(loaddataset['last_review'])
loaddataset['review_year'] = loaddataset['last_review'].apply(lambda last_review:last_review.year)
loaddataset['review_year'] = loaddataset['review_year'].fillna(0)
loaddataset['review_year'] = loaddataset.review_year.astype(int)
loaddataset = pd.concat([loaddataset[(loaddataset['availability_365']==0) & (loaddataset['review_year']==2019)], loaddataset[loaddataset['availability_365']>0]])
```

```python
loaddataset['review_per_month'] = loaddataset['reviews_per_month'].fillna(0)
```

```python
fig = sns.catplot(x='neighbourhood_group', y='availability_365', kind='box', hue='room_type',data=loaddataset, palette='bright')
fig.fig.suptitle('Dimana Letak Area yang Paling Populer', fontsize=15, y=1.05)
fig.savefig("popular area.png", bbox_inches = 'tight')
```

```python
fig = sns.catplot(x='neighbourhood_group', y='price', data=loaddataset,kind='bar', hue='room_type', palette='bright')
fig.fig.suptitle('di mana letak daerah yang paling mahal?', fontsize=15, y=1.05)
fig.savefig("price area.png", bbox_inches = 'tight')
```

```python
import plotly.express as px

lat = np.mean(loaddataset['latitude'])
lon = np.mean(loaddataset['longitude'])

fig = px.density_mapbox(loaddataset,lat ='latitude', lon="longitude",
                      radius=2, center=dict(lat=lat, lon=lon), zoom=10,
                       mapbox_style="carto-positron")
fig.show()
```

```python
fig = sns.catplot(x='number_of_reviews', y='neighbourhood_group', kind='violin', hue='room_type',data=loaddataset, palette='bright')
fig.fig.suptitle('Dimana Letak Daerah yang memiliki review paling banyak?', fontsize=15, y=1.05)
fig.savefig("Neighbour group.png", bbox_inches = 'tight')
```
